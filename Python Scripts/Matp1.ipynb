```python
import matplotlib.pyplot as plt
import random as r
# para no abrir ventana con la gr√°fica
%matplotlib inline 
```

```python
def LCG(seed, n, a=1664525, c=1013904223, m=2**16):
    numbers = []
    for i in range(n):
        seed = (a * seed + c) % m
        numbers.append(seed/m)

    return (numbers)
LCG(5,1000)
```

```python
u = LCG(5, 1000) #asigno a u la lista de n√∫meros aleatorios uniformes 

plt.figure()
plt.hist(u)

plt.show()
```

```python
# n = Cantidad de numeros
# semilla = valor inicial y no se incluye en la lista generada

def LCG(semilla, n, a=5, b=7, c=16):
    numbers = []
    for i in range(n):
        semilla = (a * semilla + b) % c
        numbers.append(semilla/c)

    return numbers

#se llama a la funcion con 1 de semilla y que genere 10 numeros
LCG(2, 10)
```
